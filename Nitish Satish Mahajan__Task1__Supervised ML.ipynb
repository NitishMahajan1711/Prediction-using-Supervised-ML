# Importing Packages

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn import metrics
```

# Reading Data From Remote Link

```python
url="http://bit.ly/w-data"
data=pd.read_csv(url)
data1=data
print("The data is imported successfully")
data
```

# Plotting the distribution of scores

```python
data.plot(x='Hours',y='Scores',style='o',color='r')
plt.title('Hours vs Percentage Score')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

# Splitting training and testing data

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,1].values
x_train, x_test, y_train, y_test= train_test_split(x, y,train_size=0.80,test_size=0.20,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
linearRegressor= LinearRegression()
linearRegressor.fit(x_train, y_train)
y_predict= linearRegressor.predict(x_train)
```

```python
regressor = LinearRegression()  
regressor.fit(x_train, y_train) 

print("Training complete.")
```

# Plotting the regression line

```python
line = regressor.coef_*x+regressor.intercept_
```

# Plotting for the test data

```python
plt.scatter(x, y)
plt.plot(x, line);
plt.show()
```

```python
print(x_test) 
y_predict = linearRegressor.predict(x_test)
```

# Test Vs Training Score

```python
print('Test Score')
print(regressor.score(x_test, y_test))
print('Training Score')
print(regressor.score(x_train, y_train))
```

```python
data= pd.DataFrame({'Actual': y_test,'Predicted': y_predict[:5]})
data
```

# Checking the efficiency of model

```python
mean_squ_error = mean_squared_error(y_test, y_predict)
mean_abs_error = mean_absolute_error(y_test, y_predict)
print("Mean Squred Error:",mean_squ_error)
print("Mean absolute Error:",mean_abs_error)
```

```python
print('Mean Absolute Error: ',metrics.mean_absolute_error(y_test,y_predict))
print('Root Mean Squared Error Error: ',np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```
